```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Pervichnyj analiz dannyh (EDA)
df_train = pd.read_csv("smartphone_battery_drain_dataset.csv")

print("--- 1. PERVICHNYJ ANALIZ (EDA) ---")
print(f"Ishodnyj razmer dataset-a: {df_train.shape}")
print("\nPervye 5 strok:")
display(df_train.head(5))

print("\nStolbcy i tipy dannyh:")
df_train.info()

print("\nStatisticheskoe opisanie:")
display(df_train.describe())
```

## 2. Очитска, обработка

```python
df = df_train.drop_duplicates().copy()
print(f"Kolichestvo najdennyh dublikatov: {df_train.duplicated().sum()}")

print("\nPropushchennye znacheniya:")
print(df.isnull().sum())

# Statistika dlya obosnovaniya udaleniya propuskov
stats = df.groupby("Usage_Mode")[
    ["CPU_Usage_%", "Battery_Temperature_C"]
].agg(["count", "mean", "median", "std"])
print("\nStatistika po gruppam dlya analiza propuskov:")
display(stats)

df.dropna(subset=["CPU_Usage_%", "Battery_Temperature_C"], inplace=True)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

df_before = df.copy()
df_before["Data_State"] = "Before"

# Korrekciya znachenij
df["Battery_Temperature_C"] = df["Battery_Temperature_C"].clip(lower=0,upper=60)
for pct_col in ["CPU_Usage_%", "Brightness_Level_%"]:
    df[pct_col] = df[pct_col].clip(0, 100)

df["RAM_Usage_MB"] = df["RAM_Usage_MB"].clip(lower=0)
df["Screen_On_Time_min"] = df["Screen_On_Time_min"].clip(lower=0)
df["Battery_Drop_Per_Hour"] = df["Battery_Drop_Per_Hour"].clip(lower=0)

df_after = df.copy()
df_after["Data_State"] = "After"
df_compare = pd.concat([df_before, df_after])

plt.figure(figsize=(18, 12))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df_compare, x="Data_State", y=col, palette="Set2")
    plt.title(f"Comparison: {col}")
    plt.xlabel("")

plt.tight_layout()
plt.show()
```

## 3. Кодирование и создание новых столбцов

```python
print("--- 3. KODIROVANIE I NOVYE PRIZNAKI ---")

df = pd.get_dummies(df, columns=["Network_Type", "Usage_Mode"], prefix=["Net", "Mode"])
df["Charging_State"] = df["Charging_State"].map({"Charging": 1, "Not Charging": 0})

freq_encoding = df["App_Running"].value_counts().to_dict()
df["App_Running_freq"] = df["App_Running"].map(freq_encoding)
df.drop("App_Running", axis=1, inplace=True)

# Sozdanie novyh stolbcov
df["Intensity"] = (df["CPU_Usage_%"] * df["RAM_Usage_MB"]) / 1000
df["Total_Battery_Drain"] = (df["Screen_On_Time_min"] / 60) * df["Battery_Drop_Per_Hour"]
temp_threshold = df["Battery_Temperature_C"].median()
df["High_Temperature"] = (df["Battery_Temperature_C"] > temp_threshold).astype(int)

print(f"Sozdany priznaki: Intensity, Total_Battery_Drain, High_Temperature")
print(f"Itogovaya razmernost: {df.shape}")
display(df.head(5))

df.to_csv("smartphone_battery_processed.csv", index=False)
print("\nFAJL SOHRANEN: LAB_2/smartphone_battery_processed.csv")
```
